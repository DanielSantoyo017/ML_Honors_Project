```python
#Importing Python libs
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Creating weekly visits df holding weekly student visit counts to the Tut center form 18-23
Weekly_Visits = pd.read_csv('Semester_Wk_Level_Data_18_23.csv')
```

```python
display(Weekly_Visits)
```

```python
 #Creating X and Y vairables with Weekly_Vists df, where x is the semester weeks 
# and y is the number of students who visited for the week

Wk_X = Weekly_Visits.iloc[:,0]

Vs_Y = Weekly_Visits.iloc[:,1]
```

```python
#specifiying scatter plot style
plt.style.use('seaborn')
```

```python
#Scatter plot with x and y

plt.scatter(Wk_X,Vs_Y)
```

```python
#Creating Daily visits df holding Daily student visit counts to the Tut center form 18-23
Daily_Visits = pd.read_csv('Day_Level_Data_18_23.csv')
```

```python
display(Daily_Visits)
```

```python
 #Creating X and Y vairables with Daily_Vists df, where x is the Days of year (1-365)
# and y is the number of students who visited for the day

Day_X = Daily_Visits.iloc[:,0]

Dvs_Y = Daily_Visits.iloc[:,1]
```

```python
#Scatter plot with x and y

plt.scatter(Day_X,Dvs_Y)
```

```python
#importing sklearn libs
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
#Option 1 - Reshaping to (-1,1)
Wk_X_RS = np.array(Wk_X).reshape(-1,1)
```

```python
#Creating Linear regression for the weekly visit data
reg = LinearRegression()

#fitting LR to Weekly data
reg.fit(Wk_X_RS,Vs_Y)

# Generating x and y values for the LR
X_vals = Wk_X_RS
y_vals = reg.predict(X_vals)
```

```python
#ploting Linear regression atempt
plt.scatter(Wk_X,Vs_Y)
plt.plot(X_vals,y_vals, color = "r")
plt.show()
```

```python
#Importing Polynomial features from sklearn
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Assinging Poly Features
poly_features = PolynomialFeatures(degree = 230, include_bias=False)

X_poly = poly_features.fit_transform(Wk_X_RS)

#Creating polynomial regression for the weekly visit data
reg = LinearRegression()

#fitting LR to Weekly data
reg.fit(X_poly,Vs_Y)

# Generating x and y values for the LR
X_vals = Wk_X_RS
X_vals_poly = poly_features.transform(X_vals)

y_vals = reg.predict(X_vals_poly)

#ploting Poly regression atempt
plt.scatter(Wk_X,Vs_Y)
plt.plot(X_vals,y_vals, color = "r")
plt.show()
```
